```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Univariate Analysis

# Count Plot for Categorical Columns

```python
for col in df:
    if df[col].dtype == 'O':
        sns.countplot(x=col,data=df)
        plt.show()
```

# Distribution of Numerical columns

```python
for col in df:
    if df[col].dtype != 'O':
        sns.histplot(df[col],kde=True)
        plt.show()
```

# Pricing Range of Each Category

```python
sns.boxplot(x='category',y='price',data=df)
plt.show()
```

# Profit Margin to each Category

```python
sns.boxplot(x='category',y='profit_margin',data=df)
plt.show()
```

# Delivery Time Required for each Category

```python
sns.boxplot(x='category',y='delivery_days',data=df)
plt.show()
```

# Discount Percentage Range to Each Category

```python
sns.boxplot(x='category',y='discount_percentage',data=df)
plt.show()
```

# Category Wise Revenue Comparison

```python
sns.barplot(x='category',y='revenue',data=df)
plt.show()
```

# Multivariate analysis

# Colors availability in each Category

```python
crosstab = pd.crosstab(df['category'], df['color'])
ax = crosstab.plot(kind='bar', figsize=(10, 5))
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1))
plt.xticks(rotation=0)
plt.show()
```

# Category availability in Online vs Retail

```python
crosstab = pd.crosstab(df['category'], df['store_type'])
ax = crosstab.plot(kind='bar', figsize=(10, 5))
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1))
plt.xticks(rotation=0)
plt.show()
```

# Category availability in Diffrent Material Types

```python
crosstab = pd.crosstab(df['category'], df['material'])
ax = crosstab.plot(kind='bar', figsize=(10, 5))
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1))
plt.xticks(rotation=0)
plt.show()
```

# Store Type availability by each Brand

```python
crosstab = pd.crosstab(df['brand'], df['store_type'])
ax = crosstab.plot(kind='bar', figsize=(10, 5))
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1))
plt.xticks(rotation=0)
plt.show()
```

# Ratio of Materials Used by Each Brand for Manufacturing Products

```python
crosstab = pd.crosstab(df['brand'], df['material'])
ax = crosstab.plot(kind='bar', figsize=(10, 5))
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1))
plt.xticks(rotation=0)
plt.show()
```

# Revenue of Diffrent Materials in Diffrent Seasons

```python
pivot_table = df.pivot_table(index='season', columns='material', values='revenue', aggfunc='sum')
pivot_table.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Seasonal Revenue by Material')
plt.xlabel('Season')
plt.ylabel('Revenue')
plt.legend(title='Material')
plt.grid(True)
plt.show()
```

# Seasonal Revenue of Each Brand

```python
pivot_table = df.pivot_table(index='season', columns='brand', values='revenue', aggfunc='sum')
pivot_table.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Seasonal Revenue by brand')
plt.xlabel('Season')
plt.ylabel('Revenue')
plt.legend(title='brand')
plt.grid(True)
plt.show()
```

```python
corr = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
